```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import dash
import dash_bootstrap_components as dbc
import re
```

```python
# Load Data
data = pd.read_excel("data/data.xlsx", index_col=0)
data = data[data.country != '.DS'].reset_index()
weights=pd.read_excel("data/lda_df_weights_update.xlsx", index_col=0)
df_merged = weights
```

```python
def indices(lst, element):
    cleanString = re.sub("'",'',lst)
    cleanString = re.sub(",",'', cleanString)
    cleanString=cleanString[1:len(cleanString)-1]
    cleanString=sorted(cleanString.split())
    result = []
    offset = -1
    while True:
        try:
            offset = lst.index(element, offset+1)
        except ValueError:
            return result
        result.append(offset)
```

```python
intro_words = "\t"
with open('assets/intro.txt') as this_file:
    for a in this_file.read():
        if "\n" in a:
            intro_words += "\n \t"
        else:
            intro_words += a

method_words = "\t"
with open('assets/methods.txt') as this_file:
    for a in this_file.read():
        if "\n" in a:
            method_words += "\n \t"
        else:
            method_words += a
```

```python
year = np.array(data["year"].unique())
drop_values = data.country.unique()
topic_list={"International Diplomacy":"Topic 1",
            "War and Power":"Topic 2",
            "Conflict in Africa":"Topic 3",
            "Middle Eastern Conflict":"Topic 4",
            "Climate Change":"Topic 5",
            "Communism":"Topic 6"}
topic_list2={"Topic 1":"International Diplomacy",
             "Topic 2":"War and Power",
             "Topic 3":"Conflict in Africa",
             "Topic 4":"Middle Eastern Conflict",
             "Topic 5":"Climate Change",
             "Topic 6":"Communism"}
topics = pd.read_excel("data/topic_trend_final.xlsx", index_col=0)
topics["Topic_Name"]=topics.Topic.map(topic_list2)
topic_values = topics.Topic_Name.unique()
```

```python
# Build App
app = JupyterDash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])



# styling the sidebar
SIDEBAR_STYLE = {
    "position": "fixed",
    "top": 0,
    "left": 0,
    "bottom": 0,
    "width": "16rem",
    "padding": "2rem 1rem",
    "background-color": "#f8f9fa",
    "z-index": "10",
}

# padding for the page content
CONTENT_STYLE = {
    "margin-left": "18rem",
    "margin-right": "2rem",
    "padding": "1rem 2rem",
}

sidebar = html.Div(
    [
        html.H2("Navigate", className="display-4"),
        html.Hr(),
        dbc.Nav(
            [
                dbc.NavLink("Introduction", href="/", active="exact"),
                dbc.NavLink("Naive Counter", href="/page-1", active="exact"),
                dbc.NavLink("LDA Results", href="/page-4", active="exact"),
                dbc.NavLink("General Results", href="/page-2", active="exact"),
                dbc.NavLink("Country Specific Results", href="/page-3", active="exact")

            ],
            vertical=True,
            pills=True,
        ),
    ],
    style=SIDEBAR_STYLE,
)

content = html.Div(id="page-content", children=[], style=CONTENT_STYLE)

app.layout = html.Div([
    dcc.Location(id="url"),
    sidebar,
    content
])

@app.callback(
    Output("page-content", "children"),
    [Input("url", "pathname")]
)
def render_page_content(pathname):
    if pathname == "/":
        return [
            html.H1('Modern Data Analytics Project:'),
            html.H2('Politicians and Climate Change'),
            html.Br(),
            html.Div(dcc.Markdown(intro_words)),
            html.Div('{} countries are represented in these speeches.'.format(len(data['country'].unique()))),
            html.Br(),
            html.H3('Methdology'),
            html.Div(dcc.Markdown(method_words)),
            html.Img(src=app.get_asset_url('LDA_k6_cloud.png'))
        ]
    elif pathname == "/page-1":
        return [
            html.H1("Naive Counter"),
            html.Div("A very basic way to investigate the content of these speeches is by searching for a specific term."),
            html.Div("This of course ignores any contextual differences in meaning the word might have"),
            html.Br(),
            html.Div("Enter a search term and press enter. The search takes a few seconds to process, please be patient."),
            html.Br(),
            html.Strong("Search: "),
            dcc.Input(id='term_select', type='text', debounce=True, value='people', required=True),
            html.Br(),
            html.Br(),
            html.Div(id='naive_text', children=[]), #, style={'color': 'blue', 'fontSize': 14}
            dcc.Graph(id="naive_graph", figure={})    
        ]
    elif pathname == "/page-2":
        return [
            html.H1('Overall trends'),
            html.Div("Please select any topics you wish to explore."),
            dcc.Dropdown(
                id='general_dropdown',
                options=[
                    {'label': i, 'value': i} for i in topic_values
                    ],
                multi=True,
                value=['Climate_Change']
                ),
             dcc.Graph(id='general_graph', figure={})
        ]
    elif pathname == "/page-3":
        return [
            html.H1("Country Specific Trends"),
            html.Div("This page lets you explore topics for any specific country. Please choose any country and topic you're interested in"),
            html.Br(),
            dbc.Row([
                dbc.Col(
                    dcc.Dropdown(
                        id='country_dropdown',
                        options=[{'label': i, 'value': i} for i in drop_values],
                        value = 'ALB',
                        placeholder='Select a country',
                        style={'width': '100%'}
                    )
                ),
                dbc.Col(dcc.Dropdown(
                id='topic_dropdown',
                options=[
                    {'label': i, 'value': i} for i in topic_list.keys()
                    ],
                value = 'International Diplomacy',
                 style={'width': '100%'}
                )) 
            ]),
            dcc.Graph(id='country_graph', figure={})   
            
        ]
    elif pathname == "/page-4":
        return [
            html.Iframe(src=app.get_asset_url("lda_updated.html"), style=dict(
                position="relative",
                left="300ox",
                top="0",
                width="1240px",
                height="770px",
                transform="scale(0.75)",
                border="0",
                marginTop="-100px",
                marginLeft="-150px",
                overflow=""))
        ]
#    position: relative;
#     top: 0px;
#     width: 1240px;
#     height: calc(100vh + 30px);
#     transform: scale(0.85);
#     margin-top: -50px;
#     margin-left: -95px;
#     border: 0; 
######################################################################################
################################### Naive Counter ####################################
######################################################################################

@app.callback(
    Output(component_id='naive_text', component_property='children'),
    Output(component_id='naive_graph', component_property='figure'),
    Input('term_select', 'value'))

def update_graph(term_select):
    
    data_c = data
      
    term = term_select.lower()
    count = np.empty(0)
    
    for i in data_c.index:
        x=len(indices(data_c.lemmatized.loc[i],term))
        count=np.append(count,x)
    data_c["word_freq"]=count
    
    data_year=data_c.groupby("year").sum("word_freq")
    data_year["years"]=data_year.index
    
    scatter = px.scatter(data_year,x="years",y="word_freq",trendline="lowess",
                        trendline_color_override="red")
    scatter.update_layout(
        yaxis = dict(
            showgrid = True,
            zeroline = True,
            showline = True,
            gridcolor = 'lightgrey'),
        xaxis_title="",
        yaxis_title='Mentions',
        plot_bgcolor="#fff")
    
#     container = 'Number of times "{}" was mentioned in a given year'.format(term_select)
    container = html.P(children=[
        html.Span('Number of times '),
        html.Strong(html.Em(term_select)),
        html.Span(' was mentioned in a given year')
    ])
    
    return container, scatter


######################################################################################
#################################### General Page ####################################
######################################################################################

@app.callback(
    Output(component_id='general_graph', component_property='figure'),
    [dash.dependencies.Input('general_dropdown', 'value')])

def update_output(value):
    
    data_filtered = topics[topics.Topic_Name.isin(value)]
    
    fig = px.line(data_filtered, x="year", y="prevalence", color='Topic_Name').update_layout(
    yaxis = dict(
    showgrid = True,
    zeroline = True,
    showline = True,
    gridcolor = 'lightgrey'),
    xaxis_title="",
    yaxis_title='Prevalence',
    plot_bgcolor="#fff")
    
    return fig


######################################################################################
#################################### Country Page ####################################
######################################################################################

@app.callback(
    Output(component_id='country_graph', component_property='figure'),
    Input('country_dropdown', 'value'),
    Input('topic_dropdown', 'value'))

def update_output_div(country_dropdown, topic_dropdown):
    
    data_filtered = df_merged.loc[df_merged['country'] == country_dropdown]
    year = np.array(data_filtered["year"].unique())
    weights=np.array(data_filtered[topic_list[topic_dropdown]])

    scatter = px.scatter(x=year, y=weights, trendline="lowess",
                        trendline_color_override="red")
    scatter.update_xaxes(range=[min(df_merged["year"]),max(df_merged["year"])])
    scatter.update_layout(
    yaxis = dict(
    showgrid = True,
    zeroline = True,
    showline = True,
    gridcolor = 'lightgrey'),
    xaxis_title="",
    yaxis_title='Prevalence',
    plot_bgcolor="#fff")
    
       
    return scatter
        




# Run app and display result inline in the notebook
app.run_server(mode='jupyterlab', debug=False,dev_tools_ui=False,dev_tools_props_check=False,port=8070)
```
